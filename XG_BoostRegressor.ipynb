```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset).astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_female', 'sex_male', 'smoker_no', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
!pip install xgboost
```

```python
from xgboost import XGBRegressor
regressor=XGBRegressor(n_estimators=1000,max_depth=5,eta=0.01,subsample=0.7,colsample_bytree=0.5)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
import pickle
filename="Final_Model_XGBoostReg.sav"
pickle.dump(regressor,open(filename,'wb'))
```
